## Capítulo 4

```python
def cuarange(x0):
    return x0**2 + 2*x0 -15
lim0=6
limy=50
x1=np.linspace(-lim0,lim0)

plt.figure(figsize=(16,8))
plt.grid()
plt.plot(x1, cuarange(x1),label='$x^{2}+2x-15$')
plt.xlim(-lim0,lim0)
plt.ylim(-20,10)
plt.axhline(y = 0, color = 'black', linestyle = '-')
plt.vlines(x=0, ymin=-limy, ymax=limy, colors='k', linestyles='solid')
plt.xticks(np.arange(-lim0, lim0, 1))
plt.xlabel('x', fontsize="30")
plt.ylabel('y', fontsize="30")
plt.title(r'Función $f(x)=x^{2}+2x-15$', fontsize="25")
plt.plot(-5, 0, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green",label='x=-5')
plt.plot(3, 0, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green",label='x=3')
plt.legend(loc=0, fontsize="30")
#plt.savefig('cua441.png')   
```

```python
def fracctable(x0):
    return (2*x0+4)/(x0+3)

lim0=10
x2=np.linspace(-lim0,-3.001,1000)
x3=np.linspace(-2.99,lim0,1000)

plt.figure(figsize=(16,8))
plt.grid()
plt.plot(x2,fracctable(x2),'c',label='$f(x)=\dfrac{2x+4}{x+3}$')
plt.plot(x3,fracctable(x3),'c')
plt.xlim(-lim0,lim0)
plt.ylim(-10,10)
plt.axhline(y = 0, color = 'black', linestyle = '-')
plt.vlines(x=0, ymin=-limy, ymax=limy, colors='k', linestyles='solid')
plt.xticks(np.arange(-lim0, lim0, 1))
plt.xlabel('x', fontsize="30")
plt.ylabel('y', fontsize="30")
plt.title(r'Función $f(x)=\dfrac{2x+4}{x+3}$', fontsize="25")
plt.legend(loc=0, fontsize="30")
#plt.savefig('inefracc0.png')
```

## Funció inversa

```python
def raiz3(x0):
    return np.sqrt(x0+3)
def cua3(x00):
    return x00**2 -3

x10=np.linspace(-3,5,500)
x100=np.linspace(-0,5,500)

plt.figure(figsize=(16,8))
plt.plot(x10, raiz3(x10),label=r'$\sqrt{x+3}$')
plt.plot(x100, cua3(x100),label=r'$x^{2}-3$')
plt.axhline(y = 0, color = 'black', linestyle = '-')
plt.vlines(x=0, ymin=-6, ymax=8, colors='k', linestyles='solid')
plt.ylim(-6,6)
plt.xlim(-3,5)
plt.grid()
plt.legend(loc=0) 
```

### Rectas

```python
def f2016(x):                               #Define la función uno
    return 5*x+60

def f2020(x):                               #Define la función dos
    return 10*x-90

x1=np.linspace(0,365)                       #Define el rango en que se graficará

plt.figure(2)                                #Define el entorno del gráfico
plt.grid()                                  #Muestra la malla en el gráfico
plt.plot(x1,f2016(x1),label=r'$f(x)=5x+60$')     #Plotea la función uno
plt.plot(x1,f2020(x1),label=r'$f(x)=10x-90$')    #Plotea la función dos
plt.legend(loc=0)                           #Ubica la legenda, loc=0 elige el mejor espacio
plt.xlabel('Días')                          #Etiqueta eje x
plt.ylabel('Mascarillas')                   #Etiqueta eje y
plt.xlim(0,365)                             #Define el limite que se mostrará en eje x
plt.title('Uso de mascarillas 2016 vs 2020')#Define el titulo del gráfico
plt.axhline(y = 0, color = 'black', linestyle = '-')
#plt.savefig('mascarillas.png')              #Guarda la imagen
```

### Asíntotas

```python
def asinvert(x):
    return (x**2 +2)/(x-2)

lim=2.5
limy=100
x30=np.linspace(-lim,1.99,1500 )
x31=np.linspace(2.001,lim+2,1500 )

plt.figure(4,figsize=(16,8))
plt.grid()
plt.plot(x30,asinvert(x30),'c','-',label=r'f(x)')
plt.plot(x31,asinvert(x31),'c','-', label='_nolegend_')

plt.axhline(y = 0, color = 'black', linestyle = '-', label='_nolegend_')
plt.vlines(x=0, ymin=-limy, ymax=limy, colors='k', linestyles='solid', label='_nolegend_')
plt.vlines(x=2, ymin=-limy, ymax=limy, colors='r', linestyles='--',label='Asíntota')
plt.title(r'$f(x)=\dfrac{x^{2}+2}{x-2}$', fontsize="30")
plt.xlim(-lim,lim+2)
plt.ylim(-limy,limy)
plt.xlabel('x', fontsize="30")
plt.ylabel('y', fontsize="30")
plt.legend(loc=0, fontsize="30")
plt.savefig('asintver.png')
```

```python
def asinh(x):
    return (8*x)/(x**2 -2*x+1)

lim=12
limy=20
x3=np.linspace(-lim,lim+2,300 )
plt.figure(5,figsize=(16,8))
plt.grid()
plt.plot(x3,asinh(x3),'-',label=r'f(x)')
plt.axhline(y = 0, color = 'r',label='Asíntota')
plt.vlines(x=0, ymin=-limy, ymax=100, colors='k', linestyles='solid')
plt.title(r'$f(x)=\dfrac{8x}{x^{2}-2x+1}$', fontsize="30")
plt.xlim(-lim,lim+2)
plt.ylim(-10,100)
plt.xlabel('x', fontsize="30")
plt.ylabel('y', fontsize="30")
plt.legend(loc=0, fontsize="30")
plt.savefig('asinth.png')
```
